```
import pandas as pd
import numpy as np
```

```
pd.set_option("display.max.columns", None)
pd.set_option("display.precision", 2)
```

```
O = ['A','A','A','A','A','A','A','A']

T = [121,123,125,127,129,131,132,133]

I = [
     '01 01 20 09:52:47','01 01 20 19:52:18','01 01 20 09:52:17','01 01 20 10:18:18',
     '01 01 20 09:51:48','01 01 20 09:51:47','01 01 20 10:19:14','01 01 20 09:54:42',
     ]
F = [
     '01 01 20 11:59:47','01 01 20 19:52:50','01 01 20 09:57:29','01 01 20 11:55:19',
     '01 01 20 09:51:50','01 01 20 09:54:29','01 01 20 10:19:33','01 01 20 09:55:59',
     ]
```

```
df = pd.DataFrame({'O': O, 'T':T, 'I':pd.to_datetime(I), 'F':pd.to_datetime(F)})#, 'D': pd.to_datetime(F)-pd.to_datetime(I)})
df.sort_values(by=['I', 'F'], inplace=True); df
```

```
df_ = pd.melt(df, id_vars=['O', 'T'], value_vars=['I', 'F'], var_name='Type', value_name='Timestamp').sort_values(by='Timestamp'); df_
```

```
t = [0]*df_.shape[0]
for i, s in enumerate(zip(df_.Type.iloc[1:], df_.Type.iloc[:], df_['T'].iloc[1:], df_['T'].iloc[:])):
    if s[0] != s[1] and s[2] == s[3]: 
        t[i] = ('UNICO')

df_ = df_.assign(TESTE = t)
df_
```

```
df.merge(df_[df_.TESTE=='UNICO'][['T', 'TESTE']], on='T')
```
