```
a  = ['isso é uma lista', 'que possui o objetivo', 'de entender', 'como funciona o explode' ]
b  = ['essa é uma outra lista', 'porém', 'essa lista', 'não irá utilizar o', 'pandas explode' ]
c  = ['humm', 'então temos 3 listas', 'essa daqui', 'só', 'vai tentar', 'ver', 'se tudo', 'funciona', 'corretamente' ]

df = pd.DataFrame(
    {
        'Text_A':[a, b, c],
        'Func':[981, 281,271], 
        'Text_B':[['texto', 'palavra', 'char'], ['char', 'index', 'var'], ['text with word', 'word with char']],
     }
      )


df
```

```
df_explode_Text_A = df.explode('Text_A', ); df_explode_Text_A.reset_index(drop=True)
```

```
# depois em Text_B
df_explode_Text_A.explode('Text_B').reset_index(drop=True)
```

# Solução sem explode

```
# 1o passo: concatenar todos os valores da coluna em um array
np.concatenate(df.Text_A.values)
```

```
# 2o passo: o index é o str.len() de cada linha com numpy repeat
df.index.repeat(df.Text_A.str.len())
```

```
#3o passo: criar um dataset que recebe as colunas do 1o passo e o index do 2o passo.
s=pd.DataFrame({'Text_A':np.concatenate(df.Text_A.values)},index=df.index.repeat(df.Text_A.str.len()))
s
```

```
# 4o passo: fazer join com as outras colunas
s.join(df.drop('Text_A', axis=1), how='left')
```

```
s=pd.DataFrame({'Text_A':np.concatenate(df.Text_A.values)},index=df.index.repeat(df.Text_A.str.len()))
s = s.join(df.drop('Text_A',1),how='left').reset_index(drop=True); s
```

## Repetir o processo para a coluna Text_B

```
s_B = pd.DataFrame({'Text_B':np.concatenate(s.Text_B.values)},index=s.index.repeat(s.Text_B.str.len())); s_B
s_B = s_B.join(s.drop('Text_B',axis=1),how='left').reset_index(drop=True); s_B
```
